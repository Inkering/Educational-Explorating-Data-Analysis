## Databases for exploring correlations

```python
def per_student(total, enrolled):
  return total/enrolled
```

```python
def race_percent(num_race, total):
  if num_race == None or total == None:
    return 0
  else:
    return float(num_race)/float(total)
```

The represented races include AM (American Indian or Alaska Native), AS (Asian), HI (Hispanic/Latino), BL (Black), WH (White), HP (Pacific Islander), and TR (Two or More Races)

All years database (8th grade)

```python
#total database for all years
revs = states_full_key[(states_full_key['TOTAL_REVENUE'] <= 5000000)][['PRIMARY_KEY','TOTAL_REVENUE','AVG_MATH_8_SCORE', 'GRADES_8_BL','GRADES_8_AS','GRADES_8_HI','GRADES_8_AM','GRADES_8_WH','GRADES_8_G']]
revs['REVENUE_PER_STUDENT'] =  list(map(per_student, revs['TOTAL_REVENUE'],revs['GRADES_8_G']))
revs['AS_PER'] =  list(map(race_percent, revs['GRADES_8_AS'],revs['GRADES_8_G']))
revs['BL_PER'] =  list(map(race_percent, revs['GRADES_8_BL'],revs['GRADES_8_G']))
revs['AM_PER'] =  list(map(race_percent, revs['GRADES_8_AM'],revs['GRADES_8_G']))
revs['WH_PER'] =  list(map(race_percent, revs['GRADES_8_WH'],revs['GRADES_8_G']))
revs['HI_PER'] =  list(map(race_percent, revs['GRADES_8_HI'],revs['GRADES_8_G']))
revs = revs.drop(columns=['TOTAL_REVENUE', 'GRADES_8_AS', 'GRADES_8_BL','GRADES_8_AM','GRADES_8_WH','GRADES_8_HI'])
revs = revs.dropna(axis=0, how="any") #just replace null values
revs
```

## Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
def linear_regression(X, y):
    w, _, _, _ = np.linalg.lstsq(X, y, rcond=-1)
    return w
```

```python
def fit_model(dfx, dfy, cols):
    # Note: this is might not really be a fair way to do a training / testing split
    # since it will result in instances that were very close in time falling in
    # both the training and testing set
    X_train, X_test, y_train, y_test = \
        train_test_split(dfx, dfy)

    model = LinearRegression()
    model.fit(X_train, y_train)
    for i, c in enumerate(cols):
      print('w_%d (the weight for %s) = %f' % (i+1, c, model.coef_[i]))
    
    print('w_%d (the weight for the bias term) = %f' % (len(cols) + 1, model.intercept_))

    print("\ny actual is within 5 of y predicted:")
    print('Accuracy on training set', (abs(model.predict(X_train) - y_train)<5).mean())
    print('Accuracy on testing set', (abs(model.predict(X_test) - y_test)<5).mean())
    print("\ny actual is within 10 of y predicted:")
    print('Accuracy on training set', (abs(model.predict(X_train) - y_train)<10).mean())
    print('Accuracy on testing set', (abs(model.predict(X_test) - y_test)<10).mean())
    return model
```
